Load and show the content of lab1.txt file

```python
import pandas

file = pandas.read_csv("/content/lab1.txt")
print(file)
```

Add a new column "z" where z = sqrt(x^2+y^2) (don't use loops!)

```python
import numpy as np
file['z'] = np.sqrt(file['x']**2 + file['y']**2)
print(file)
```

Remove the column "code"

```python
file = file.drop(['code'], axis=1)
print(file)
```

Choose only rows for which x>1000

```python
file = file.query('x>1000')
file
```

Choose only columns file, x, y i z

```python
file = file[['file', 'x', 'y', 'z']]
print(file)
```

Check statistics - how many rows for each value in "file" column

```python
file['file'].value_counts()
```

Choose only rows for which "file" = 1141 and save it to numpy array (use *values* method)

```python
file = file[file['file']==1141]
file
```

Check the shape of numpy array

```python
file.shape
```

Calculate max and average values in x and y columns

```python
np.max(file, axis=0)[1:3]
```

Create a plot of x and y columns

```python
import matplotlib.pyplot as plt

plt.figure
plt.title("Plot")
plt.plot(file['x'],file['y'])
plt.draw()
```

Create two histograms of x with 10 and 100 bins

```python
plt.figure
plt.subplot(1,2,1)
plt.title("hist. x with 10 bins")
plt.hist(file['x'],bins=10);
plt.subplot(1,2,2)
plt.title("hist. x with 100 bins")
plt.hist(file['x'],bins=100);
```
